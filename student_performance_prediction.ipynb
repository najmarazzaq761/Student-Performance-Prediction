## feature Engineering

## droping column which have no effect on target variable

```python
df.drop(["StudentID","Age","Gender","Ethnicity"],axis=1,inplace=True)
```

```python
df.shape
```

```python
x=df.drop('GradeClass', axis=1)
y=df["GradeClass"]
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=101)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## loading and training

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(multi_class='multinomial', solver='lbfgs')
logreg.fit(x_train,y_train)
```

## prediction

```python
y_pred_logreg=logreg.predict(x_test)
print(y_pred_logreg)
```

## testing

```python
probabilities=logreg.predict_proba(x_test)[:,1]
print(probabilities.shape)
print(probabilities)
```

```python
for i in range(len(y_test)):
  print(round(probabilities[i],3),"--->",y_pred_logreg[i])
```

## checking accuracy

```python
from sklearn.metrics import classification_report, accuracy_score
print("classification model")
print("-----"*40)
logreg_accuracy=round(accuracy_score(y_test,y_pred_logreg)*100,2)
print("accuracy is",logreg_accuracy,"%")
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=12)
model.fit(x_train,y_train)
```

```python
y_pred_model=model.predict(x_test)
from sklearn.metrics import classification_report, accuracy_score
print("classification model")
print("-----"*40)
model_accuracy=round(accuracy_score(y_test,y_pred_model)*100,2)
print("accuracy is",model_accuracy,"%")
```
